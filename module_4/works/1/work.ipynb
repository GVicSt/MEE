```python
import pandas as pd

piz_df=pd.read_csv('https://raw.githubusercontent.com/GVicSt/DataSets/main/pizas_data_small.csv', sep='\t')
piz_df
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression

def get_regression_output(main_index, other_indexes, data, split_p, input):
    # разделяю данные на целевые и другие
    x = data[main_index]
    y = data[other_indexes].values.reshape(-1,1)
    # разделяю данные на тренировочные и тестовые
    x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = split_p, random_state = 0)

    # использую линейную регрессию
    regressor = LinearRegression()
    regressor.fit(x_train, y_train)

    # получаю результаты
    output =      pd.DataFrame(regressor.predict(pd.DataFrame(input)))
    output_test = pd.DataFrame(regressor.predict(x_test))

    return [output.T,output_test,y_test]

new_pizza = pd.DataFrame({'размер(см)':[75.0],'популярность':[1.0],'эксклюзивность':[1.0],'масса(грамм)':[1187.0],'калории':[895.0]})

out = get_regression_output(['размер(см)','популярность','эксклюзивность','масса(грамм)','калории'],
                            'цена(руб)',
                            piz_df,
                            0.2,
                            new_pizza
                            )

print(f'цена которая нужна для соответствия качество/цена, при характеристиках:\n\n{new_pizza}\n\nравна: {out[0][0].values[0]}\n\n\
      \nданные теста:\n  предсказанные данные:\n{out[1].values}\n\n  исходные данные:\n{out[2]}')
```
