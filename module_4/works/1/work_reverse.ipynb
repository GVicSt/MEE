цель:

```python
import pandas as pd
```

```python
piz_df=pd.read_csv('https://raw.githubusercontent.com/GVicSt/DataSets/main/pizas_data_small.csv', sep='\t')
piz_df
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression

def get_regression_output(main_index, other_indexes, data, split_p, input):
    # разделяю данные на целевые и другие
    x = data[main_index].values.reshape(-1,1)
    y = data[other_indexes]
    # разделяю данные на тренировочные и тестовые
    x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = split_p, random_state = 0)

    # использую линейную регрессию
    regressor = LinearRegression()
    regressor.fit(x_train, y_train)

    # получаю результаты
    output =      pd.DataFrame(regressor.predict(pd.DataFrame([float(input)])).reshape(-1,1), index=other_indexes)
    output_test = pd.DataFrame(regressor.predict(x_test)).T
    output_test.index = other_indexes

    return [output.T,output_test.T,y_test]

out = get_regression_output('цена(руб)',
                            ['размер(см)','популярность','эксклюзивность','масса(грамм)','калории'],
                            piz_df,
                            0.2,
                            3500 # input('введите цену: ')
                            )

print(f'характеристики которые нужны для соответствия качество/цена, при цене 3500 рублей:\n{out[0]}\n\n\
      \nданные теста:\n  предсказанные данные:\n\n{out[1]}\n  исходные данные:\n\n{out[2]}')
```
