```python
def proizvol(formule):
    formule_list = formule.split('+')
    x = []
    for i in range(len(formule_list)):
        x.append(formule_list[i].split('*'))
    temp = []
    for i in range(len(formule_list)):
        if 'x' not in x[i]:
            temp.append(0)
        if '1' == x[i][0] and x[i][-1] == '1':
            temp.append(1)
        elif '1' != x[i][0] and x[i][-1] == '1':
            temp.append(str(x[i][0]))
        elif int(x[i][-1]) > 1:
            if int(x[i][-1])-1 == 1:
                temp.append(str(int(x[i][-1]) * int(x[i][0]))+'x')
            else:
                temp.append(str(int(x[i][-1]) * int(x[i][0]))+'x'+'^'+str(int(x[i][-1])-1))
    result = ''
    for i in range(len(temp)):    
        result += (str(temp[i]))
        if i != len(temp)-1:
            result += ' + '
    return result

print(proizvol('1*x**3'))
```

```python

def mat_determinant_3x3(a):
    if len(a) != 3 or len(a[0]) != 3:
        return "break"
    z0,z1,z2,z3,z4,z5 = 1,1,1,1,1,1
    for i in range(len(a)):
        z0 *= a[i][i]
        if i+1 >= 2:
            z1 *= a[i][0]
        else:
            z1 *= a[i][i+1]
        if i+1 >= 2:
            z2 *= a[0][i]
        else:
            z2 *= a[i+1][i]
        z3 *= a[2-i][i]
        if i >= 2:
            z4 *= a[2][2]
        else:
            z4 *= a[2-i-1][i]
        if i >= 2:
            z5 *= a[1][1]
        else:
            z5 *= a[2-i][i+1]
    return (z0+z1+z2)-(z3+z4+z5)
print(mat_determinant_3x3([[1,8,9],[2,6,3],[15,-1,2]]))
```

```python
def mat_determinant_3x3(a):
    if len(a) != 3 or len(a[0]) != 3:
        return "break"
    z0 = a[0][0] * a[1][1] * a[2][2]
    z1 = a[1][0] * a[0][2] * a[2][1]
    z2 = a[0][1] * a[1][2] * a[2][0]
    z3 = a[0][2] * a[1][1] * a[2][0]
    z4 = a[0][0] * a[2][1] * a[1][2]
    z5 = a[1][0] * a[0][1] * a[2][2]
    return (z0+z1+z2)-(z3+z4+z5)

print(mat_determinant_3x3([[1,8,9],[2,6,3],[15,-1,2]]))
```
