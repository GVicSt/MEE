```python
import pandas as pd

gen_df=pd.read_csv('https://raw.githubusercontent.com/GVicSt/DataSets/genshin/genshin_impact_data.csv', sep='\t')[['name','rarity','element','level','friendship']]
gen_df
```

```python
# стандартизацию данные
# правде не знаю зачем... просто так в задании написано
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaled_gen_df = scaler.fit_transform(gen_df[['rarity','level','friendship']])
scaled_gen_df
```

```python
#целевой переменной будет уровень персонажа
other = gen_df[['rarity','friendship']]
main = pd.DataFrame(gen_df['level'])
print(other.head())
print(main.head())
```

```python
# разделение на тренировочные и проверочные данные функцией train_test_split из model_selection
from sklearn.model_selection import train_test_split
main_train, main_test, other_train, other_test = train_test_split(other, main, test_size=0.3,
random_state=42)
print(main_train.shape)
print(main_test.shape)
print(other_train.shape)
print(other_test.shape)
```

```python
# использую метод машинного обучения - Логистическая регрессия
# потому что простой
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
model.fit(main_train, other_train)
z = []
for i in range(32):
    z.append(model.predict(other.iloc[[i]]).item())

pd.DataFrame([main['level'],pd.Series(z).T],index=['original','predicted']).T
```
