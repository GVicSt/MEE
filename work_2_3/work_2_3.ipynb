```python
def mat_determinant_2x2(a):
    z0,z1 = 1,1
    for i in range(len(a)):
        z0 *= a[i][i]
        z1 *= a[i][1 - i]
    return z0-z1

def mat_mul_num(a,b):
    matrix = []
    matrix_temp = []
    for i in range(len(a)):
        for j in range(len(a[0])):
            z = a[i][j] * b
        
            matrix_temp.append(z)
        matrix.append(matrix_temp)
        matrix_temp = []
    return matrix

def mat_reverse(a):
    if mat_determinant_2x2(a) == 0:
        return "break"
    matrix = []
    matrix_temp = []
    for i in range(len(a)):
        for j in range(len(a[0])):
            z = a[len(a)-i-1][len(a[0])-j-1]
            if i != j:
                z *= -1    
            matrix_temp.append(z)
        matrix.append(matrix_temp)
        matrix_temp = []
    
    result = mat_mul_num(matrix,1/mat_determinant_2x2(matrix))

    return result

matrix = [[1,8],[2,6]]
print(mat_reverse(matrix))
```
