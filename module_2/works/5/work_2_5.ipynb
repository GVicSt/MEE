```python
def mat_make(width,height):
    matrix = []
    for i in range(width):
        matrix_temp = []
        for j in range(height):
            matrix_temp.append(0)
        matrix.append(matrix_temp)
    return matrix

def mat_trasposition(a):
    matrix = []
    matrix_temp = []
    for i in range(len(a)):
        for j in range(len(a[0])):
            z = a[j][i]
            matrix_temp.append(z)
        matrix.append(matrix_temp)
        matrix_temp = []
    return matrix

def mat_determinant_at(mat):
  det = 0
  m = len(mat)
  if m == 2:
    det = mat[0][0] * mat[1][1] - mat[0][1] * mat[1][0]
  if m > 2:
    for k in range(m):
      det += mat[0][k] * (-1) ** k *\
          mat_determinant_at([mat[i][:k] + mat[i][k+1:] for i in range(1,m)])
  return det

def mat_mul_num(a,b):
    matrix = []
    matrix_temp = []
    for i in range(len(a)):
        for j in range(len(a[0])):
            z = a[i][j] * b
            matrix_temp.append(z)
        matrix.append(matrix_temp)
        matrix_temp = []
    return matrix

def mat_inverse(mat):
  det_mat = mat_determinant_at(mat)
  if det_mat == 0:
    return 0
  else:
    m = len(mat)
    join_mat = [[mat[1][1], (-1) * mat[1][0]], [(-1) * mat[0][1], mat[0][0]]]
    inv_mat = mat_mul_num(mat_trasposition(join_mat),round((1 / det_mat),2))
  return inv_mat

def mat_mul_mat(a,b):
    matrix = []
    for x in a:
        matrix_temp = []
        for y in zip(*b):
            z = 0
            for i,j in zip(x, y):
                z += i*j
            matrix_temp.append(z)
        matrix.append(matrix_temp)
    return matrix

def ab(x,y):
    z = mat_make(1,2)
    w = mat_make(2,2)
    h = mat_make(1,2)
    
    for i in x:
        w[0][1] += i
    for i in y:
        z[0][1] += i
    for i in x:
        w[0][0] += i**2
    for i in range(len(x)):
        z[0][0] += x[i] * y[i]
    w[1][1] = len(x)
    w[1][0] = w[0][1]
    h = mat_mul_mat(z,mat_inverse(w))

    print('A =', w)
    print('B =', z)
    print('X =', h)

    return h

def get_pred(z,y):
   for i in range(z):
      print(i,':',y[0][0]*i+y[0][1])

level = [1,2,3,4,5]
frags = [3,6,5,7,10]
    
g = ab(level, frags)
get_pred(int(input()), g)
```

```python
def mat_make(width,height):
    matrix = []
    for i in range(width):
        matrix_temp = []
        for j in range(height):
            matrix_temp.append(0)
        matrix.append(matrix_temp)
    return matrix

def mat_determinant_at(mat):
  det = 0
  m = len(mat)
  if m == 2:
    det = mat[0][0] * mat[1][1] - mat[0][1] * mat[1][0]
  if m > 2:
    for k in range(m):
      det += mat[0][k] * (-1) ** k * mat_determinant_at([mat[i][:k] + mat[i][k+1:] for i in range(1,m)])
  return det

def abc(x,y):
    b, out = mat_make(1,3), mat_make(1,3)
    a, a1, a2, a3 = mat_make(3,3),mat_make(3,3),mat_make(3,3),mat_make(3,3)
    
    for i in x:
        a[0][0] += i**4
        a[0][1] += i**3
        a[0][2] += i**2
        a[1][2] += i
    a[1][0], a[1][1], a[2][0], a[2][1], a[2][2] = a[0][1], a[0][2], a[0][2], a[1][2], len(x)

    for i in range(len(x)):
        b[0][0] += x[i]**2 * y[i]
        b[0][1] += x[i] * y[i]
        b[0][2] += y[i]

    for i in range(len(a)):
        for j in range(len(a[0])):
            if i == 0:
                a1[i][j] = b[0][j]
            else:
                a1[i][j] = a[i][j]
            if i == 1:
                a2[i][j] = b[0][j]
            else:
                a2[i][j] = a[i][j]
            if i == 2:
                a3[i][j] = b[0][j]
            else:
                a3[i][j] = a[i][j]

    out[0][0] = mat_determinant_at(a1)/mat_determinant_at(a)
    out[0][1] = mat_determinant_at(a2)/mat_determinant_at(a)
    out[0][2] = mat_determinant_at(a3)/mat_determinant_at(a)

    print('A =', a)
    print('A1 =', a1)
    print('A2 =', a2)
    print('A3 =', a3)
    print('B =', b)
    print('X =', out)

    return out[0]

def get_pred(z,y):
   for i in range(1,z+1):
      print(i,':',y[0]*i**2+y[1]*i+y[2])

level = [1,2,3,4,5]
frags = [5,15,25,45,65]

n_abc = abc(level,frags)
get_pred(int(input()),n_abc)
```
